```python
def show_predict(model, x, y):
    X0 = np.linspace(min(x), max(x), 100)
    Y0 = model.predict(X0)
    plt.figure(figsize=(10, 6))
    plt.grid(alpha=0.3)
    plt.scatter(x, y)
    plt.plot(X0, Y0, 'r', linewidth=4)
    plt.show()
def get_metrics(model, X_train, x_test, Y_train, y_test, show=True,return_MAE=False):
    R2_train = round(model.score(X_train, Y_train), 4)
    MAE_train = round(mean_absolute_error(model.predict(X_train),Y_train), 4)
    R2_test = round(model.score(x_test, y_test), 4)
    MAE_test = round(mean_absolute_error(model.predict(x_test), y_test),4)
    
    if show:
        print(' Обучающие данные:')
        print(f' R^2 score = {R2_train}')
        print(f' MAE = {MAE_train}')
        print(' Тестовые данные:')
        print(f' R^2 score = {R2_test}')
        print(f' MAE = {MAE_test}')
    if return_MAE:
        return(R2_train, MAE_train, R2_test, MAE_train)
    else:
        return(R2_train, R2_test)
for x0 in x:
    X_train, x_test, Y_train, y_test = train_test_split(x[[x0]].values,y, train_size=0.8, random_state=42)
    model = LinearRegression().fit(X_train, Y_train)
    print(f'Парная модель для фактора {x0}:')
    get_metrics(model, X_train, x_test, Y_train, y_test, show=True)
    show_predict(model, x[[x0]].values, y)
    print('\n' * 3)
```
